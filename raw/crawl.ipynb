```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import numpy as np
import pandas as pd
import time
```

```python
# driver = r'chromedriver.exe'
driver = ChromeDriverManager().install()
```

```python
import csv
```

```python
href_list = []
data_list = []
urllist = []
N = 2161
flag = 0
starturl = 'https://vnexpress.net/no-luc-lua-chon-cap-pho-trong-tham-lang-cua-ong-trump-4741840.html'
wd = webdriver.Chrome(driver)
wd.get(starturl)
wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(0.5)
data = wd.find_elements(By.CLASS_NAME, 'item-tag')
print(f'Khởi tạo được {len(data)} tags từ link trên: ', end='\t')
for d in data:
    print(d.text, end='\t')
    urllist.append((d.text,d.find_element(By.TAG_NAME, 'a').get_attribute('href')))
print()
wd.execute_script("window.open('');")
wd.switch_to.window(wd.window_handles[0])
# wd2 = webdriver.Chrome(driver)
for _, url in urllist:
    p = url.split('-')[-1]
    while p!='p21':
        try:
            wd.get(url)
        except:
            continue
        wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)
        data = wd.find_elements(By.CLASS_NAME, 'title-news')
        hrefs = []
        for d in data:
            try:
                href = d.find_element(By.TAG_NAME, 'a').get_attribute('href')
            except:
                continue
            finally:
                hrefs.append(href)
        try:
            next = wd.find_element(By.CLASS_NAME, 'next-page')
        except:
            break       
        
        for href in hrefs:
            if len(data_list)>= N:
                flag = 1
                break
            domain = href.split('/')[2] # loại bỏ link quảng cáo
            if href not in href_list and domain =='vnexpress.net':
                href_list.append(href)
                wd.switch_to.window(wd.window_handles[1])
                try:
                    wd.get(href)
                except:
                    continue
                wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(1)
                try:
                    tags = wd.find_elements(By.CLASS_NAME, 'item-tag')
                    ps = wd.find_elements(By.CLASS_NAME, 'Normal')
                except:
                    continue
                data_ = '\n'.join([p.text for p in ps])
                if len(tags)<=0:
                    continue
                
                tag_list = []
                for tag in tags:
                    try:
                        tag_list.append(tag.text)
                        t = (tag.text,tag.find_element(By.TAG_NAME, 'a').get_attribute('href'))
                        if t not in urllist:
                            urllist.append(t)
                    except:
                        continue
                print(len(data_list)+1, end='\t')
                print(href)
                data_list.append((data_, ', '.join(tag_list)))
        if len(data_list)>= N:
            flag = 1
            break
        wd.switch_to.window(wd.window_handles[0])                     
        url = next.get_attribute('href')
        p = url.split('-')[-1]
    print('Total data: ',len(data), end=',\t')
    print('Total link: ',len(href_list), end=',\t')
    print('total crawl: ',len(data_list))
    # mỗi 50 data lưu 1 lần
    with open("data.csv","w") as f:
        wr = csv.writer(f)
        for i,j in data_list:
            wr.writerow([i,j])
    with open("tag.csv","w") as f:
        wr = csv.writer(f,delimiter=',')
        for i,j in urllist:
            wr.writerow([i,j])
    print('Đã ghi nhận: ',len(data_list))
    if flag==1:
        break
# wd2.close()
wd.close()
```
